#Introduction

---

Within rule-based approaches for the design of chemical libraries, the synthesis schemes and available resources-based design have shown the importance of the knowledge of smart reactions in the fields of the chemical space exploration. Integrating this knowledge has allowed us to design synthetically accessible libraries, like BI-CLAM developed by Boehringer Ingelheim (Lessel et al, 2009) and CHIPMUNK (Humbeck et al., 2018). Another example is the TIN library, it is a combinatorial chemical database built from the synthetic accessibility of multi-component reactions that has more than 28 million structures of products that are innovative and synthetically accessible (Dorschner et al., 2011). In a similar way, and considering the click chemical reactions of triazoles, a ZINClick database was generated, made from more than 16 million of 1, 2, 3- and 1, 4- triazoles, whose structures are innovative and synthetically accessible (Massarotti et al., 2014). Another open access virtual library is 'Screenable Chemical Universe Based on Intuitive Data OrganizatioOn' (SCUBIDOO) (Chevillard & Kolb, 2015). In SCUBIDOO, 58 solid reactions were applied to 18 561 molecular construction blocks to generate more than 21 million compounds. Another virtual library is the REAL database, described in the VirtualFlow platform that has more than 1 400 million on demand compounds.




Even thought each library has its variations, common steps can be distinguished which consist of:  

**1.**	Choose a synthesis approach to be followed and the reactions to be used (e.g. biology-oriented synthesis (BIOS), diversity-oriented synthesis (DOS), traditional synthesis or combinatorial chemistry).

**2.**	Build or identify the building blocks to be used (generally commercially available blocks).

**3.**	Identify the characteristics of the building blocks for the approach to be followed. In this step you can filter the fragments according to functional groups or price.

**4.**	Establish the reactions to be used in a legible format for the computer (e.g. SMIRKS)

**5.**	Use of chemoinformatics tools for the automation of the synthesis approach.

**6.**	Implementation of filters for the exclusion and removal of duplicated compounds or compounds with structural alerts.
Next, we will describe an example for the design of chemical libraries using a chemical reactions-based approach.

## *2. Compound enumeration using RDKit*

Design of a library of bis-heterocycle compounds obtained through click chemistry using Python and the `RDKit` package.

### 1. Selection of a synthesis approach.
This example is based on the approach reported by Shafi et al.  to obtain bis-heterocycle compounds, linking building blocks of 5-membered heterocycles that have one or two heteroatoms (at least one nitrogen, oxygen or sulfur) to a set of building blocks that have azide through the formation of a disubstituted 1,4 ,2,3-triazole by click chemistrY (Shafi et al., 2012).

###2. Building or identification of building blocks
Building blocks used in this example were taken from the Sigma Aldrich catalog through ZINC DB, that has 124 368 building blocks. The duplicated compounds were deleted, and we selected the building blocks that fulfill Congreve’s “Rule of thirds” (RO3). The selected database can be found as [*Sigma_bb.sdf*](https://github.com/DIFACQUIM/Cursos/blob/main/Datasets/Sigma_bb.sdf.zip) .

**NOTE**: Visit the following link to see other building blocks catalogs available on ZINC: https://zinc.docking.org/catalogs/

### 3. Identify the characteristics of the building blocks for the approach to be followed.
The SMARTS were built to detect the presence of appropriate functional groups to do the approach described in number one.

* 5-membered heterocyclic ring with one (N, O or S) or two heteroatoms (N, O, S; at least one N) and a nucleophilic substituent (- OH, –SH, –NH2 ).

###5. Use of chemoinformatics tools for the automatization of the synthesis approach.
In this example we will use Python and the  `RDKit` package to generate the bis-heterocycle compound library.

Examples for the building of chemical libraries using other chemoinformatics tools such as KNIME and DataWarrior can be found in:  [Saldivar-González et al. 2020](https://jcheminf.biomedcentral.com/articles/10.1186/s13321-020-00466-z).

a) Install packages and modules.

```python
!pip install rdkit
```

```python
import pandas as pd
import rdkit as rk
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.Chem.rdMolDescriptors import CalcNumHeteroatoms
from rdkit.Chem import Draw
from rdkit.Chem import PandasTools
from rdkit.Chem import rdMolDescriptors
```

b) Read Sigma's construction blocks library using the `Supplier`RDKit function.

```python
supp = Chem.SDMolSupplier('/content/Sigma_bb.sdf')
for mol in supp:
    if mol is not None: mol.GetNumAtoms()
```

c) Create a list that contains Sigma construction blocks.


```python
mols = [x for x in supp]
len(mols) #Number of construction blocks
```

d) Filter construction blocks according to the previous established SMARTS.

```python
#SMARTS 5-membered heterocyclic ring
patt1= Chem.MolFromSmarts('[$([NX3;H2;!$(NC=O)]),$([#16X2H]),$([OX2H])]-[cr5;$([cr5]:1:[nr5,or5,sr5]:[cr5]:[cr5]:[nr5,or5,sr5]:1),$([cr5]:1:[cr5]:[nr5,or5,sr5]:[cr5]:[cr5]:1)]')
het5 = [x for x in mols if x.HasSubstructMatch(patt1)]
from rdkit.Chem import Draw
Draw.MolsToGridImage(list(het5), subImgSize=(300,300))
```

```python
##SMARTS Terminal 3-bromo alkyne or substituted chlorine
patt2= Chem.MolFromSmarts('[Br,Cl][#6]C#[CH1]')
alkynes = [x for x in mols if x.HasSubstructMatch(patt2)]
Draw.MolsToGridImage(list(alkynes), subImgSize=(150,150))
```

```python
#SMARTS Azide
patt3= Chem.MolFromSmarts('[N;H0;$(N-[#6]);D2]=[N;D2]=[N;D1]')
azide = [x for x in mols if x.HasSubstructMatch(patt3)]
Draw.MolsToGridImage(list(azide), subImgSize=(300,300))
```

e) Establish the coupling reactions and enumerate the compounds

```python
#Nucleophilic substitution

rxn=AllChem.ReactionFromSmarts('[#6;a;r5:1]-[$([NX3;H2;!$(NC=O)]),$([#16X2H]),$([OX2H]):2].[#35,#17]-[#6:3][C:4]#[C:5]>>[#6;a;r5:1]-[$([NX3;H]),$([#16X2]),$([OX2]):2]-[#6:3][C:4]#[C:5]')
prods1 = AllChem.EnumerateLibraryFromReaction(rxn,[het5,alkynes])
smis = list(set([Chem.MolToSmiles(x[0],isomericSmiles=True) for x in prods1]))
```

```python
#Visualization of the reaction scheme using SMIRKS
rxn
```

```python
print("Number of generated compunds:", len(smis))
```

```python
#Click chemistry reaction

rxn2= AllChem.ReactionFromSmarts('[#6:7][C:6]#[CH1:5].[#6:4]-[#7:3]=[N+]=[#7-]>>[#6:4]-[#7:3]-1-[#6:5]=[#6:6](-[#6:7])-[#7]=[#7]-1 |c:10,t:7|')
prods2 = AllChem.EnumerateLibraryFromReaction(rxn2,[[ Chem.MolFromSmiles(x) for x in smis ],azide])
smis2 = list(set([Chem.MolToSmiles(x[0],isomericSmiles=True) for x in prods2]))
```

```python
#Visualization of the reaction scheme using SMIRKS
rxn2
```

```python
print("Number of generated compunds:", len(smis2))
```

```python
#Visualize SMILES of the generated compounds
smis2
```

f) Export results in a *csv* format.

```python
#Export results as .CSV File
df = pd.DataFrame(smis2, columns=["colummn"])
df.to_csv('bis_heterocycles.csv', index=False)
df
```

g) Visualize the generated compounds

```python
PandasTools.AddMoleculeColumnToFrame(df,'colummn','Molecule',includeFingerprints=True)
print([str(x) for x in  df.columns])

PandasTools.FrameToGridImage(df,column= 'Molecule', molsPerRow=5,subImgSize=(350,350))
```
