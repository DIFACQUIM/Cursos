# Introduction

A molecular representation is a codification of a chemical structure in a way that is comprehensible for a software. Most of the representations have a specific purposes. For example:
* Communication and quantitative analysis.
* Data storage.
* Structure searching.


#*1. Common molecular representations*
---

The most common graphic description of compounds is the bidimensional (2D). Nowadays there is plenty of software that aids in the drawing of chemical structures, storage, and interconversion between different file formats. Some of this software have free academic versions like [MarvinSketch](https://chemaxon.com/marvin)  and [ACD/ChemSketch](https://www.acdlabs.com/resources/free-chemistry-software-apps/chemsketch-freeware/), and some others are commercial like [ChemDraw](https://chemdrawdirect.perkinelmer.cloud/js/sample/index.html), [Schr√∂dinger](https://www.schrodinger.com/) and [MOE](), by mentioning  a few. Three-dimensional structures (3D) are also widely used, in molecular modelling. These representations provide powerful and intuitive tools for the comprehension of various chemistry aspects. However, these representations have limitations, especially while performing daily Chemoinformatics tasks. This is why lineal notations (1D) were developed, these save up memory space, are useful for storage, search and management of data; as well as the efficient processing of large amounts of compounds (thousands and millions).

#*2. [**Linear representations**](https://doi.org/10.1186/s13321-020-00466-z)*
---


In Chemoinformatics, the molecular information is usually represented by linear notations. The most common linear notations used are:  SMILES, SMARTS, InChI and InChIkeys.

 > [**SMILES**](https://doi.org/10.1021/ci00057a005)





*Simplified Molecular input Line System* (SMILES) captures the structure of a molecule accurately in strings of alphanumeric characters. They allow efficient storage and speed up the processing of a large number of molecules.

The notation is ruled by the following:


1.   The atoms are represented by their atomic symbol. The hydrogens of saturated atoms are not explicitly represented.

image.png

2.   The neighbors atoms are placed next to one another. The bonds are represented as simple (-), double (=), triple (#) or aromatic (:), usually the simple and aromatic bonds are omitted.

image.png

3. Place in parentheses specific ramifications of the molecular structure.

image.png


4. For the lineal representation of ring structures, a bond "breaks" in each ring and the atoms that connect it must have on one side the same digit in the textual representation.

image.png
5. The atoms in aromatic rings are represented by lowercase letters. Sometimes there are problems for the aromaticity perception.

image.png


Even though SMILES notation is a clear way to describe chemical structures, multiple SMILES representations exist for the same molecule (visible in the ring coding). This is why canonic SMILES were generated; to ensure the existence of a singular representation of a compound in a database or identify duplicated compounds, this depends on the software that is used. Therefore, it is advisable to generate the SMILES using only one program.

>[**SMARTS**](https://www.daylight.com/dayhtml/doc/theory/theory.smarts.html)

*SMILES Arbitrary Target specification* (SMARTS) is a language developed to specify structural patterns used to relate molecules and reactions. It uses rules that are extensions of the SMILES. However, the labels for the atoms and the bonds also include logical operators, and other symbols that allow it to be more inclusive.
Their main applications have been searching for molecules with a particular substructure and filtering molecules with substructures associated with promiscuity and toxicological problems.

The SMARTS language provides multiple symbols that describe atomic properties and bonds beyond those used in the SMILES notation

Since chemical patterns representations are relatively new, the amount of interfaces where the user can graphically create patterns is limited. Some examples of editors for managing SMARTS notation are MarvinSketch, JSME, SMARTeditor and the web editor PubChem's Sketcher.

Another problem when using SMARTS is that there are a lot of written patterns that match additional molecules (false positives), and they can overlook others (false negatives). False positives are a problem when we want to apply transformations (e.g., SMIRKS) to matching molecules, since they can give as a result exotic valence and charge states that were not intentional.



>[**InChI Keys**](https://jcheminf.biomedcentral.com/articles/10.1186/s13321-015-0068-4)

*International Chemical IDentifier* or InChI, was developed by the IUPAC with the goal of establishing a unique label for each compound and ease its storage. It resolves SMILES ambiguities, but its harder for humans to interpret them.

The InChl Keys format is an identifier with a fixed length of 27 characters. The first part represents the molecular constitution, the second part the position of the hydrogens, stereochemistry, isotopes, metallic ligands adn finally it indicates the protonation of the compound.

**Visualize in spheres form**

```python
def smi2conf(smiles):
    #Convert SMILES to rdkit.Mol with 3D coordinates
    mol = Chem.MolFromSmiles(smiles)
    if mol is not None:
        mol = Chem.AddHs(mol)   #Add hydrogens
        AllChem.EmbedMolecule(mol) #Generate conformations
        AllChem.MMFFOptimizeMolecule(mol, maxIters=200) #Force field
        return mol
    else:
        return None
```

```python
smiles = 'COc3nc(OCc2ccc(C#N)c(c1ccc(C(=O)O)cc1)c2P(=O)(O)O)ccc3C[NH2+]CC(I)NC(=O)C(F)(Cl)Br'
```

```python
smi2conf(smiles)
```

```python
def MolTo3DView(mol, size=(300, 300), style="stick", surface=False, opacity=0.5):
    assert style in ('line', 'stick', 'sphere', 'carton')
    mblock = Chem.MolToMolBlock(mol)
    viewer = py3Dmol.view(width=size[0], height=size[1])
    viewer.addModel(mblock, 'mol')
    viewer.setStyle({style:{}})
    if surface:
        viewer.addSurface(py3Dmol.SAS, {'opacity': opacity})
    viewer.zoomTo()
    return viewer
```

```python
conf = smi2conf(smiles)
viewer = MolTo3DView(conf, size=(600, 300), style='sphere')
viewer.show()
```

## 3.9. Compare compounds with SMARTS

The compariison of compounds is useful for the analysis of structural analogs and in the process of drug optimization. For more information, visit the following reference [*ACS Omega* 2019, 4, 1, 1027‚Äì1032](https://pubs.acs.org/doi/10.1021/acsomega.8b03390)

1. Import the folowing tools

```python
from rdkit.Chem import rdFMCS
from rdkit.Chem.Draw import rdDepictor
```

2. To help with the visualization use the first function then define the size of the image.

```python
rdDepictor.SetPreferCoordGen(True)
IPythonConsole.drawOptions.minFontSize=25
```

2. Define the molecule we are going to compare.

```python
mol1 = Chem.MolFromSmiles('CN1CCC23C4C1CC5=C2C(=C(C=C5)O)OC3C(C=C4)O')
mol2 = Chem.MolFromSmiles('CC(=O)OC1C=CC2C3CC4=C5C2(C1OC5=C(C=C4)OC(=O)C)CCN3C')

Draw.MolsToGridImage([mol1, mol2])
```

3. Define the function that highlights the differences between both molecules. In this case, first we define Maximum Common Substructure (MCS) and then we highlight the differences.

**Note:** To know more about the MCS algorithm you can reviwe the following reference: J Comput Aided Mol Des. 2002 Jul; 16(7):521-33
Dalke A, Hastings J., FMCS: a novel algorithm for the multiple MCS problem. J. Cheminf. 2013; 5 (Suppl 1): O6

Raymond JW., Willet P., Maximum common subgraph isomorphism algorithms for the matching of chemical structures. J Comput Aided Mol Des. 2002 Jul; 16(7):521-33

```python
mcs = rdFMCS.FindMCS([mol1,mol2])
print("MCS SMARTS string:", mcs.smartsString)
```

```python
m1 = Chem.MolFromSmarts(mcs.smartsString)
Draw.MolToImage(m1, legend="MCS1")
```

```python
def view_difference(mol1, mol2):
    mcs = rdFMCS.FindMCS([mol1,mol2])
    mcs_mol = Chem.MolFromSmarts(mcs.smartsString)
    match1 = mol1.GetSubstructMatch(mcs_mol)
    target_atm1 = []
    for atom in mol1.GetAtoms():
        if atom.GetIdx() not in match1:
            target_atm1.append(atom.GetIdx())
    match2 = mol2.GetSubstructMatch(mcs_mol)
    target_atm2 = []
    for atom in mol2.GetAtoms():
        if atom.GetIdx() not in match2:
            target_atm2.append(atom.GetIdx())
    return Draw.MolsToGridImage([mol1, mol2],highlightAtomLists=[target_atm1, target_atm2])
```

4. View differences

```python
view_difference(mol1,mol2)
```

The differences are highlighted in red.
